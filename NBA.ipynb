```r
# Cargar librerías necesarias
library(ggplot2)  # Para las gráficas
library(dplyr)    # Para manipulación de datos

# Cargar los datos
players_stats <- read.csv("players_stats(3).csv")

# i) Representar gráficamente la distribución de las alturas (Height)
ggplot(players_stats, aes(x = Height)) +
  geom_histogram(binwidth = 1, fill = "blue", color = "black") +
  labs(title = "Distribución de las Alturas de los Jugadores (2014-2015)", 
       x = "Altura (cm)", 
       y = "Frecuencia") +
  theme_minimal()

# ii) Comparar gráficamente las alturas de los jugadores por posición
ggplot(players_stats, aes(x = Pos, y = Height, fill = Pos)) +
  geom_boxplot() +
  labs(title = "Comparación de las Alturas por Posición (2014-2015)",
       x = "Posición",
       y = "Altura (cm)") +
  theme_minimal() +
  scale_fill_brewer(palette = "Set3")

# iii) Analizar si los jugadores más altos capturan más rebotes
ggplot(players_stats, aes(x = Height, y = REB)) +
  geom_point(color = "darkgreen") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Relación entre Altura y Rebotes Capturados (2014-2015)",
       x = "Altura (cm)",
       y = "Rebotes por juego") +
  theme_minimal()

# iv) Detectar outliers en la relación Altura vs Rebotes
ggplot(players_stats, aes(x = Height, y = REB)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  geom_text(aes(label = ifelse(REB > quantile(REB, 0.95), as.character(players_stats$Player), "")), 
            vjust = -0.5, color = "darkred") +
  labs(title = "Detección de Outliers: Altura vs Rebotes (2014-2015)",
       x = "Altura (cm)",
       y = "Rebotes por juego") +
  theme_minimal()

# v) Comentario de las gráficas
# - La primera gráfica muestra que la mayoría de los jugadores tienen alturas entre 75 y 85 pulgadas (aproximadamente entre 1.90 y 2.15 metros).
# - En la segunda gráfica (boxplot) se observa que las posiciones de Pívot (C) y Ala-Pívot (PF) tienden a ser más altas, mientras que los Bases (PG) son los más bajos.
# - La tercera gráfica sugiere una correlación positiva entre altura y rebotes capturados, aunque algunos jugadores más bajos también capturan muchos rebotes.
# - En la cuarta gráfica se identifican algunos jugadores que son outliers en cuanto a la cantidad de rebotes capturados. Estos jugadores pueden destacarse por su estilo de juego (jugadores más pequeños que capturan muchos rebotes) o pueden ser casos extremos en la muestra.
```
